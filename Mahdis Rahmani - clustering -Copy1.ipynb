# Make blobs

```python
n_classes = 5 # n = 5 labels

data, labels, centers = make_blobs(n_samples = 3000,
                                    centers = n_classes,
                                    random_state = 100,
                                    return_centers = True)

# n_features = 2 as default for 2D plotting
```

```python
labels
```

```python
data[:10]
```

```python
centers
```

```python
x = []
y = []
for i in range(1000):
    x.append(data[i][0])
    y.append(data[i][1])
```

```python
plt.scatter(x, y)
plt.title("blobs sample", size =20)
```

```python
blobs = pd.DataFrame(data =  data,
                     columns = ['x', 'y'])
```

```python
blobs['labels'] = labels
```

```python
blobs.head(20)
```

```python
colors = ['red', 'orange', 'blue', 'green', 'pink']
classes = list(blobs.labels.unique())
for i in range(len(classes)): #no. of classes:
    x = blobs[blobs['labels'] == classes[i]]
    plt.scatter(x[blobs.columns[0]], x[blobs.columns[1]],
               c = colors[i],
               label = classes[i])
    
for i in range(len(classes)):
    plt.scatter(centers[i][0], centers[i][1], c = 'black' )
plt.title('blobs', size = 20)
plt.legend()
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans_model_blobs = KMeans(n_clusters = 5,
                     n_init = 12,
                     verbose = 1,
                     random_state = 0)
```

```python
kmeans_model_blobs.fit(blobs.drop(columns = ['labels']))
```

```python
predicted_labels_blobs = kmeans_model_blobs.labels_
```

```python
blobs['predicted labels'] = predicted_labels_blobs
```

```python
blobs
```

```python
colors = ['red', 'orange', 'blue', 'green', 'pink']
classes = list(blobs.labels.unique())
for i in range(len(classes)): #no. of classes:
    x = blobs[blobs['predicted labels'] == classes[i]]
    plt.scatter(x[blobs.columns[0]], x[blobs.columns[1]],
               c = colors[i],
               label = classes[i])
    
for i in range(len(classes)):
    plt.scatter(kmeans_model_blobs.cluster_centers_[i][0], kmeans_model_blobs.cluster_centers_[i][1], c = 'black' )
plt.title('KMeans - blobs', size = 20)
plt.legend()
```

```python
wrong = blobs[blobs['labels'] != blobs['predicted labels']]
```
